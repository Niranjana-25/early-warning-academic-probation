```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = r"../Dataset/Cleaned_Academic_Probation_Project.xlsx"

df = pd.read_excel(
    file,
    sheet_name="Master_Education_Data (2)"
)

print("Dataset Loaded Successfully")
print(df.shape)
```

```python
print(df.head())
```

```python
print("Average GPA:")
print(round(df['GPA'].mean(), 2))

print("\nAverage Attendance:")
print(round(df['Attendance_Percentage'].mean(), 2))

print("\nAverage Study Hours:")
print(round(df['Study_Hours_Per_Week'].mean(), 2))

print("\nAverage LMS Login Count:")
print(round(df['LMS_Login_Count'].mean(), 2))
```

```python
dept_gpa = (
    df.groupby('Department')['GPA']
    .mean()
    .sort_values(ascending=False)
)

print(dept_gpa)
```

```python
plt.figure(figsize=(8,4))

dept_gpa.sort_values().plot(
    kind='barh',
    edgecolor='black'
)

plt.title(
    "Average GPA by Department",
    fontsize=18,
    fontweight='bold',
    pad=15
)

plt.xlabel(
    "Average GPA",
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    "Department",
    fontsize=13,
    fontweight='bold'
)

plt.grid(
    axis='x',
    linestyle='--',
    alpha=0.6
)

plt.tight_layout()

plt.savefig(
    "Output_Charts/Department_GPA.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
dept_attendance = (
    df.groupby('Department')['Attendance_Percentage']
    .mean()
    .sort_values(ascending=False)
)

print(dept_attendance)
```

```python
plt.figure(figsize=(8,4))

dept_attendance.sort_values().plot(
    kind='barh',
    edgecolor='black'
)

plt.title(
    "Average Attendance by Department",
    fontsize=18,
    fontweight='bold',
    pad=15
)

plt.xlabel(
    "Attendance Percentage",
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    "Department",
    fontsize=13,
    fontweight='bold'
)

plt.grid(
    axis='x',
    linestyle='--',
    alpha=0.6
)

plt.tight_layout()

plt.savefig(
    "Output_Charts/Department_Attendance.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
probation_count = (
    df[df['Probation_Status'] == 'Yes']
    .groupby('Department')
    .size()
    .sort_values(ascending=False)
)

print(probation_count)
```

```python
plt.figure(figsize=(8,4))

probation_count.sort_values().plot(
    kind='barh',
    edgecolor='black'
)

plt.title(
    "Students on Academic Probation by Department",
    fontsize=18,
    fontweight='bold',
    pad=15
)

plt.xlabel(
    "Number of Students",
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    "Department",
    fontsize=13,
    fontweight='bold'
)

plt.grid(
    axis='x',
    linestyle='--',
    alpha=0.6
)

plt.tight_layout()

plt.savefig(
    "Output_Charts/Probation_By_Department.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,4))

sns.scatterplot(
    data=df,
    x='Backlog_Count',
    y='GPA',
    alpha=0.5,
    s=80
)

sns.regplot(
    data=df,
    x='Backlog_Count',
    y='GPA',
    scatter=False,
    color='red',
    line_kws={'linewidth':2}
)

plt.title(
    "Relationship Between GPA and Backlog Count",
    fontsize=18,
    fontweight='bold',
    pad=15
)

plt.xlabel(
    "Backlog Count",
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    "GPA",
    fontsize=13,
    fontweight='bold'
)

plt.grid(
    linestyle='--',
    alpha=0.6
)

plt.tight_layout()

plt.savefig(
    "Output_Charts/GPA_vs_Backlog.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```
