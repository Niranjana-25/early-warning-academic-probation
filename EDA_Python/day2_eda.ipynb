```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file = r"../Dataset/Cleaned_Academic_Probation_Project.xlsx"

df = pd.read_excel(
    file,
    sheet_name="Master_Education_Data (2)"
)

print("Dataset Loaded Successfully")
print(df.shape)
```

```python
print(df.columns.tolist())
df.head()
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Attendance_Percentage'], bins=20)

plt.title("Attendance Distribution")
plt.xlabel("Attendance Percentage")
plt.ylabel("Number of Students")

plt.savefig("Output_Charts/Attendance_Histogram.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['GPA'], bins=20)

plt.title("GPA Distribution")
plt.xlabel("GPA")
plt.ylabel("Number of Students")

plt.savefig("Output_Charts/GPA_Histogram.png")

plt.show()
```

```python
plt.figure(figsize=(7,7))

df['Probation_Status'].value_counts().plot.pie(
    autopct='%1.1f%%'
)

plt.title("Probation Status Distribution")
plt.ylabel("")

plt.savefig("Output_Charts/Probation_PieChart.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['Attendance_Percentage'],
    df['GPA']
)

plt.xlabel("Attendance Percentage")
plt.ylabel("GPA")
plt.title("Attendance vs GPA")

plt.savefig("Output_Charts/Attendance_vs_GPA.png")

plt.show()
```

```python
numeric_cols = [
    'Attendance_Percentage',
    'GPA',
    'Assignment_Score',
    'Quiz_Score',
    'Internal_Marks',
    'LMS_Login_Count',
    'Study_Hours_Per_Week',
    'Backlog_Count'
]

plt.figure(figsize=(10,8))

sns.heatmap(
    df[numeric_cols].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.savefig("Output_Charts/Correlation_Heatmap.png")

plt.show()
```
