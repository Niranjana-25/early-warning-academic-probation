```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file = r"../Dataset/Cleaned_Academic_Probation_Project.xlsx"

df = pd.read_excel(
    file,
    sheet_name="Master_Education_Data (2)"
)

print("Dataset Loaded Successfully")
print(df.shape)
```

```python
print(df.columns.tolist())
df.head()
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8,4))

sns.histplot(
    df['Attendance_Percentage'],
    bins=20,
    kde=True,
    edgecolor='black'
)

plt.title(
    "Attendance Distribution",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel("Attendance Percentage")
plt.ylabel("Number of Students")

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.5
)

plt.tight_layout()

plt.savefig(
    "Output_Charts/Attendance_Histogram.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.histplot(
    df['GPA'],
    bins=20,
    kde=True,
    edgecolor='black'
)

plt.title(
    "GPA Distribution",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel("GPA")
plt.ylabel("Number of Students")

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.5
)

plt.tight_layout()

plt.savefig(
    "Output_Charts/GPA_Histogram.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
plt.figure(figsize=(8,4))

ax = sns.countplot(
    data=df,
    x='Probation_Status'
)

plt.title(
    "Probation Status Distribution",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel("Probation Status")
plt.ylabel("Number of Students")

for container in ax.containers:
    ax.bar_label(container)

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.5
)

plt.tight_layout()

plt.savefig(
    "Output_Charts/Probation_Status_Distribution.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.scatterplot(
    data=df,
    x='Attendance_Percentage',
    y='GPA',
    hue='Probation_Status',
    alpha=0.7
)

plt.title(
    "Attendance vs GPA",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel("Attendance Percentage")
plt.ylabel("GPA")

plt.grid(
    linestyle='--',
    alpha=0.5
)

plt.tight_layout()

plt.savefig(
    "Output_Charts/Attendance_vs_GPA.png",
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
numeric_cols = [
    'Attendance_Percentage',
    'GPA',
    'Assignment_Score',
    'Quiz_Score',
    'Internal_Marks',
    'LMS_Login_Count',
    'Study_Hours_Per_Week',
    'Backlog_Count'
]

plt.figure(figsize=(10,8))

sns.heatmap(
    df[numeric_cols].corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.savefig("Output_Charts/Correlation_Heatmap.png")

plt.show()
```
