```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
file = r"../Dataset/Cleaned_Academic_Probation_Project.xlsx"

df = pd.read_excel(
    file,
    sheet_name="Master_Education_Data (2)"
)

df.head()
```

```python
total_students = len(df)

avg_gpa = df["GPA"].mean()

avg_attendance = df["Attendance_Percentage"].mean()

avg_assignment = df["Assignment_Score"].mean()

avg_quiz = df["Quiz_Score"].mean()

avg_internal = df["Internal_Marks"].mean()

avg_lms = df["LMS_Login_Count"].mean()

avg_study = df["Study_Hours_Per_Week"].mean()

avg_backlog = df["Backlog_Count"].mean()

probation_yes = (df["Probation_Status"] == "Yes").sum()

probation_no = (df["Probation_Status"] == "No").sum()
```

```python
print("="*50)

print("STUDENT PERFORMANCE KPIs")

print("="*50)

print(f"Total Students            : {total_students}")

print(f"Average GPA               : {avg_gpa:.2f}")

print(f"Average Attendance        : {avg_attendance:.2f}%")

print(f"Average Assignment Score  : {avg_assignment:.2f}")

print(f"Average Quiz Score        : {avg_quiz:.2f}")

print(f"Average Internal Marks    : {avg_internal:.2f}")

print(f"Average Study Hours       : {avg_study:.2f}")

print(f"Average LMS Logins        : {avg_lms:.2f}")

print(f"Average Backlogs          : {avg_backlog:.2f}")

print(f"Students on Probation     : {probation_yes}")

print(f"Students not on Probation : {probation_no}")
```

```python
department_kpi = (
    df.groupby("Department")
      .agg(
          Average_GPA=("GPA","mean"),
          Average_Attendance=("Attendance_Percentage","mean"),
          Average_Study_Hours=("Study_Hours_Per_Week","mean"),
          Average_LMS_Login=("LMS_Login_Count","mean"),
          Students=("Student_ID","count")
      )
      .round(2)
)

department_kpi
```

```python
def risk_level(row):
    if (
        row["Attendance_Percentage"] < 50 or
        row["GPA"] < 5 or
        row["Backlog_Count"] >= 3
    ):
        return "Critical Risk"

    elif (
        row["Attendance_Percentage"] < 75 or
        row["GPA"] < 6.5
    ):
        return "At Risk"

    else:
        return "Normal"

df["Risk_Level"] = df.apply(risk_level, axis=1)
```

```python
risk_summary = (
    df["Risk_Level"]
      .value_counts()
      .reset_index()
)

risk_summary.columns = ["Risk Level","Students"]

risk_summary
```

```python
probation_summary = (
    df["Probation_Status"]
      .value_counts()
      .reset_index()
)

probation_summary.columns = ["Probation Status", "Students"]

probation_summary
```

```python
department_kpi.to_csv(
    "Output_Charts/Department_KPI_Table.csv"
)

risk_summary.to_csv(
    "Output_Charts/Risk_Level_Summary.csv",
    index=False
)

probation_summary.to_csv(
    "Output_Charts/Probation_Summary.csv",
    index=False
)
```
